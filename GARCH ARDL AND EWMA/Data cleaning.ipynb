```python
import pandas as pd
import os
from functools import reduce

# Get the list of XLSX files in the directory
directory = 'francinemomo-attachments/'
xlsx_files = [file for file in os.listdir(directory) if file.endswith('.xlsx')]

# Create an empty list to store the data frames
xlsx_data = []
#del xlsx_files[2]
del xlsx_files[8]

# Loop through each XLSX file and read it into a data frame
for file in xlsx_files:
    df = pd.read_excel(directory + file)
    xlsx_data.append(df)
```

```python
#Data=pd.read_excel("DATA.xlsx")
# Merge the data frames based on the common column "Dates"
merged_df = reduce(lambda df1, df2: pd.merge(df1, df2, on='Dates', how='outer'), xlsx_data)
```

```python
# Eliminating missing values in the data
# Eliminating rows with more than 9 missing values
merged_df1 = merged_df.dropna(thresh=merged_df1.shape[1]-8)

# Display the number of eliminated rows
eliminated_rows = merged_df.shape[0] - merged_df1.shape[0]
print("Eliminated rows:", eliminated_rows)
```
